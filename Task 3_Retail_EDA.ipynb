# Sales vs Profit 

```python
sns.relplot(x = "Sales", y = "Profit", hue = "Category", data = filteredData)
```

```python
sns.catplot("Region", hue = "Sub-Category", data = filteredData, kind = "count", aspect = 1.5, palette = "Set1")
```

Its pretty evident that in  each segment, the subcategory count is being highlighted and we can find where sales and profit need to be improved 

```python
plt.figure(figsize=[22,10])
ax = sns.barplot(x="State", y="Profit", data=filteredData, palette="Paired")
plt.xticks(rotation=90, fontsize=20)
plt.yticks(fontsize=20)
plt.title("Profit earned in different states",fontsize=24)
plt.xlabel("States",fontsize=20)
plt.ylabel("Profit",fontsize=20)
plt.tight_layout()
```

Profit numbers across different states

The States with negative Profit or Loss are the weakest areas of business- Ohio, Pennsylvania, Illinois, North Carolina, Colorado, Tennessee, Arizona, Florida, Oregon, Texas.

```python
filteredData[['Sub-Category','Profit']].groupby(['Sub-Category'],as_index=True).sum().sort_values(by='Profit',ascending=True)
```

```python
s_p=filteredData.groupby("Sub-Category")["Sales","Profit"].agg(["sum"])
s_p.plot.bar(width=1, figsize=(12,8))
plt.title("Profit and Sales w.r.t subcategory", fontsize=15)
plt.xlabel("Sub-Category")
plt.ylabel("Profit and Sales")
```

Profit and Sales with respect to each Sub-Category

Tables, Bookcases, Supplies product/ Sub-Category yields loss w.r.t profit.
