# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import tree
```

# Loading raw data

```python
rawData = datasets.load_iris()
x=pd.DataFrame(rawData.data, columns=rawData.feature_names) 
```

# Studying the raw data

```python
x.head()
```

```python
x.tail()
```

```python
y=pd.DataFrame(rawData.target,columns=["Species"])
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# Creating decision trees

```python
dt= DecisionTreeClassifier() # uses gini
```

```python
dt.fit(x_train,y_train)
```

```python
dt.predict(x_test)
```

```python
y_test
```

# Checking the decision trees

```python
dt.score(x_test,y_test)
```

# Score: 1 (with gini)

```python
dt2= DecisionTreeClassifier(criterion='entropy',max_depth=5) # uses entropy
```

```python
dt2.fit(x_train,y_train)
```

```python
dt2.score(x_test,y_test)
```

# Score: 1 (with entropy)

# Prediction on testing data

```python
Y_pred=dt.predict(x_test)
```

# Visualize the trees

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dt, 
                   feature_names=rawData.feature_names,  
                   class_names=rawData.target_names,
                   filled=True)
```
